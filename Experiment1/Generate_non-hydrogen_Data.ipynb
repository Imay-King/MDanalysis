```python
import MDAnalysis as mda
import MDAnalysis.coordinates.DCD
import MDAnalysis.coordinates.DCD as dcd
import numpy as np
import pandas as pd
```

```python
topology='ala5_min_conv.pdb'
trajectory_1='ala5_first_trajectory.dcd'
trajectory_2='ala5_second_trajectory.dcd'
universe_1=mda.Universe(topology,trajectory_1)
universe_2=mda.Universe(topology,trajectory_2)
```

```python
data_1=[]
data_2=[]
```

```python
for time_step in universe_1.trajectory:
    non_hydrogen_data_1=universe_1.select_atoms('not name H*')
    data_pos_1=non_hydrogen_data_1.positions.flatten()
    #print(universe.trajectory.time)
    data1=np.append(data_pos_1,universe_1.trajectory.time) #combine positions and time together
    print(data1)
    data_1.append(list(data1))  #convert to list format and append data
```

```python
for time_step in universe_2.trajectory:
    non_hydrogen_data_2=universe_2.select_atoms('not name H*')
    data_pos_2=non_hydrogen_data_2.positions.flatten()
    #print(universe.trajectory.time)
    data2=np.append(data_pos_2,universe_2.trajectory.time) #combine positions and time together
    print(data2)
    data_2.append(list(data2))  #convert to list format and append data
```

```python
data_new_1=np.array(data_1)
data_new_2=np.array(data_2)
```

#### Concatenate numerical positions and label class

```python
final_data_1=np.concatenate((data_new_1,np.ones((data_new_1.shape[0],1))),axis=1)
final_data_2=np.concatenate((data_new_2,np.zeros((data_new_2.shape[0],1))),axis=1)
```

```python
final_data_1
```

```python
df_1=pd.DataFrame(final_data_1)
df_2=pd.DataFrame(final_data_2)
```

```python
df_1.to_csv('first_tr.csv',float_format='%.4f',header=None,index=None)
df_2.to_csv('second_tr.csv',float_format='%.4f',header=None,index=None)
```

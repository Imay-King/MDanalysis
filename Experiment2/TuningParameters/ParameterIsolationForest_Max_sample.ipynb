```python
import MDAnalysis.analysis.rms
import matplotlib.pyplot as plt
```

```python
import MDAnalysis as mda
import MDAnalysis.coordinates.DCD
import MDAnalysis.coordinates.DCD as dcd
import numpy as np
import pandas as pd
```

```python
topology='ala13_mini_conv.pdb'
trajectory='ala13_md.dcd'
universe=mda.Universe(topology,trajectory)
```

```python
trajectory_res='ala13_md_res_ca.dcd'
universe_res=mda.Universe(topology,trajectory_res)
```

```python
R = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='not name H*')
R.run()
```

```python
R_res = MDAnalysis.analysis.rms.RMSD(universe_res,
                                 select='not name H*')
R_res.run()
```

### Using max_value=50  in method

```python
rmsd_res = R_res.rmsd.T   # transpose makes it easier for plotting
time_res=rmsd_res[1][::5]
rmsd = R.rmsd.T   # transpose makes it easier for plotting
time=rmsd[1][::5]
font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 16,
         }
frames=np.arange(len(time))
neg1=[6, 8, 86, 87, 88, 91, 92, 93, 95, 98, 99, 100, 101, 104, 158, 159, 214, 247, 267, 268, 270, 272, 276, 279, 280, 281, 282, 283, 284, 285, 286, 287, 288, 289, 290, 291, 292, 293, 294, 295, 296, 297, 298, 299, 300, 301, 302, 303, 304, 305, 306, 307, 308, 309, 310, 311, 312, 313, 314, 315, 316, 317, 320, 321, 322, 329, 330, 331, 338, 339, 398, 399, 400, 401, 402, 403, 405, 407, 410, 462]
figure, ax = plt.subplots(figsize=(15,15))
plot_original,=plt.plot(frames, rmsd[2][::5], 'b--',  label="Original Frames")
plot_res,=plt.plot(frames, rmsd_res[2][::5], 'g--',  label="Restrained Frames")
for i in neg1:
    plt.plot(i,rmsd[2][::5][i],'o',c='red')
plt.legend(loc="best")
plt.xlabel("Frames",font1)
plt.ylabel(r"RMSD ($\AA$)",font1)
plt.tick_params(labelsize=16)
labels = ax.get_xticklabels() + ax.get_yticklabels()
# print labels
[label.set_fontname('Times New Roman') for label in labels]
my_x_ticks = np.arange(0, 501, 100)
legend = plt.legend(handles=[plot_original,plot_res],prop=font1)
#legend1 = plt.legend(handles=[plot_original,plot_res], prop=font1)
#legend2 = plt.legend(handles=plot_res, prop=font1)
plt.xticks(my_x_ticks)
#fig.savefig("rmsd_default_one-classSVM.jpg")
#plt.title("RMSD Plot for default hyperparameter",font2)
plt.show()
```

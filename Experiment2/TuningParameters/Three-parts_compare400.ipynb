```python
import MDAnalysis as mda
import MDAnalysis.coordinates.DCD
import MDAnalysis.coordinates.DCD as dcd
import numpy as np
import pandas as pd
```

```python
topology='ala13_mini_conv.pdb'
trajectory='ala13_md.dcd'
universe=mda.Universe(topology,trajectory)
```

```python
import MDAnalysis.analysis.rms
import matplotlib.pyplot as plt
```

```python
R1 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 1:4 and not name H*')
R1.run()
R2 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 5:9 and not name H*')
R2.run()
R3 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 10:13 and not name H*')
R3.run()
```

```python
rmsd1 = R1.rmsd.T   # transpose makes it easier for plotting
time=rmsd1[1][::5]
frames=np.arange(len(time))
neg1=[6, 99, 100, 101, 104, 113, 158, 159, 160, 214, 230, 247, 270, 272, 276, 284, 291, 294, 296, 298, 299, 300, 301, 302, 304, 308, 309, 311, 322, 338, 339, 399, 402, 403, 410, 454, 462, 477]
plt.figure(figsize=(25,15))
plt.plot(frames, rmsd1[2][::5], 'b--',  label="original frames")
for i in neg1:
    plt.plot(i,rmsd1[2][::5][i],'o',c='red')
plt.legend(loc="best")
plt.xlabel("Frames")
plt.ylabel(r"RMSD ($\AA$)")

my_x_ticks = np.arange(0, 500, 10)

plt.xticks(my_x_ticks)
plt.show()
```

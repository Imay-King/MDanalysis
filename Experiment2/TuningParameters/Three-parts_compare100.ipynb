```python
import MDAnalysis as mda
import MDAnalysis.coordinates.DCD
import MDAnalysis.coordinates.DCD as dcd
import numpy as np
import pandas as pd
import MDAnalysis.analysis.rms
import matplotlib.pyplot as plt
```

```python
topology='ala13_mini_conv.pdb'
trajectory='ala13_md.dcd'
universe=mda.Universe(topology,trajectory)
```

```python
trajectory_res='ala13_md_res_ca.dcd'
universe_res=mda.Universe(topology,trajectory_res)
```

```python
R1 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 1:4 and not name H*')
R1.run()
R2 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 5:9 and not name H*')
R2.run()
R3 = MDAnalysis.analysis.rms.RMSD(universe,
                                 select='resid 10:13 and not name H*')
R3.run()
R1_res = MDAnalysis.analysis.rms.RMSD(universe_res,
                                 select='resid 1:4 and not name H*')
R1_res.run()
R2_res = MDAnalysis.analysis.rms.RMSD(universe_res,
                                 select='resid 5:9 and not name H*')
R2_res.run()
R3_res = MDAnalysis.analysis.rms.RMSD(universe_res,
                                 select='resid 10:13 and not name H*')
R3_res.run()
```

```python
font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 16,
         }
rmsd1_res = R1_res.rmsd.T   # transpose makes it easier for plotting
time_res=rmsd1_res[1][::5]
rmsd1 = R1.rmsd.T   # transpose makes it easier for plotting
time=rmsd1[1][::5]
frames=np.arange(len(time))
neg1=[6, 99, 100, 101, 104, 113, 158, 159, 160, 214, 230, 247, 256, 268, 270, 272, 276, 284, 287, 291, 292, 293, 294, 296, 297, 298, 299, 300, 301, 302, 304, 306, 307, 308, 309, 310, 311, 313, 322, 338, 339, 399, 402, 403, 410, 411, 454, 462, 477, 493]
figure, ax = plt.subplots(figsize=(15,15))
plot_original,=plt.plot(frames, rmsd1[2][::5], 'b--',  label="Original Frames")
plot_res,=plt.plot(frames, rmsd1_res[2][::5], 'g--',  label="Restrained Frames")
for i in neg1:
    plt.plot(i,rmsd1[2][::5][i],'o',c='red')
plt.legend(loc="best")
plt.xlabel("Frames",font1)
plt.ylabel(r"RMSD ($\AA$)",font1)
plt.tick_params(labelsize=16)
labels = ax.get_xticklabels() + ax.get_yticklabels()
# print labels
[label.set_fontname('Times New Roman') for label in labels]
my_x_ticks = np.arange(0, 501, 100)
legend = plt.legend(handles=[plot_original,plot_res],prop=font1)
plt.xticks(my_x_ticks)
plt.show()
```

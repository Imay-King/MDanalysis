```python
import MDAnalysis as mda
import MDAnalysis.coordinates.DCD
import MDAnalysis.coordinates.DCD as dcd
import numpy as np
import pandas as pd
```

```python
topology='ala13_mini_conv.pdb'
trajectory='ala13_md_res_ca.dcd'
universe=mda.Universe(topology,trajectory)
```

```python
data=[]
ts=[]
```

```python
for time_step in universe.trajectory[::5]:
    non_hydrogen_data=universe.select_atoms('resid 10:13 and not name H*')
    
    data_pos=non_hydrogen_data.positions.flatten()
    #print(universe.trajectory.time)
    data1=np.append(data_pos,universe.trajectory.time)
    print(data1)
    data.append(list(data1))
```

```python
data_new=np.array(data)
```

```python
data_new
```

```python
final_data=np.concatenate((data_new,np.ones((data_new.shape[0],1),dtype=np.int64)),axis=1)
```

```python
final_data
```

```python
df=pd.DataFrame(final_data)
```

```python
df.iloc[:,-1]=df.iloc[:,-1].astype(int)
```

```python
df.iloc[:,:-2]
```

```python
df.to_csv('train_X_res_ca_10-13.csv',float_format='%.8f',header=None,index=None)
```
